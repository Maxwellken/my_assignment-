```python
#loading the built-in tips dataset
tips = sns.load_dataset('tips')
tips.head(10)
```

## Barplot using seaborn

```python
## Barplot using seaborn
sns.barplot(x='day', y='tip', data=tips)
plt.title('days for tips')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, hue='sex', palette='autumn')
plt.show()
```

```python
sns.barplot(x='size' , y='total_bill', data=tips, hue='smoker', palette='autumn')
plt.xticks(rotation=90)
plt.show()
```

## Distplot and Histplot chart

```python
 #using random numbers to plot dist plot
num = np.random.randn(100)
num
```

```python
sns.distplot(num)
plt.show()
```

```python
sns.histplot(data=num)
plt.grid(True)
plt.show()
```

## strip plot 

It shows individual data points as dots along a single axis.

Useful for small to moderate-sized datasets, where you want to see each data point, not just aggregates.

```python
sns.stripplot(x =tips['tip'])
plt.show()
```

```python
tips.head(7)
```

```python
sns.stripplot(y =tips['tip'], x= tips['smoker'], palette='winter_r')
plt.show()
```

```python
sns.stripplot(y=tips['total_bill'], x=tips['sex'], order=['Female', 'Male'])
plt.show()
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'])
plt.show()
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips, hue='smoker', palette='summer')
plt.show()
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips, hue='smoker', jitter=True)
plt.show()
```

```python
sns.stripplot(x=tips['day'], y=tips['total_bill'], data=tips, hue='sex', palette='spring')
plt.show()
```

```python
#see the distribution of male and female in the data
sns.stripplot(x='sex', y='tip', data=tips, palette='spring')
plt.show()
```

## PairGrid function

PairGrid is a flexible, customizable plotting grid in Seaborn. 
It lets you specify different plot types for the upper triangle, lower triangle, and diagonal of a grid of subplots.



```python
iris = sns.load_dataset('iris')
iris.head(10)
```

```python
x =sns.PairGrid(iris)
x= x.map(plt.scatter)
plt.show()
```

## violin  Plot

```python
sns.violinplot(x=tips['day'], y=tips['time'], palette='summer')
plt.show()
```
