```python
from bs4 import BeautifulSoup  #helps in web scrapping
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
soup.find('div')
```

```python
#find table with class named wikitable sortable
soup.find('table', class_='wikitable sortable')
```

```python
#take the second table using find_all()
table = soup.find_all('table')[1]
table
```

```python
#get the table headers   #n/b: table header will be the column name on the dataframe
world_title = table.find_all('th')
world_title
```

```python
#obtain only the table titles in the list
world_table_titles =[title.text.strip() for title in world_title]

world_table_titles
```

```python
#importing relevant libraries for dataframing
import pandas as pd, numpy as np
```

```python
#put the table Headers in a dataframe as the column names
df = pd.DataFrame(columns= world_table_titles)

df
```

```python
#follow same means to get table rows
column_data = table.find_all('tr')
column_data
```

we need to get rid of the header columns away from the table data

```python
frame = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    frame.append(individual_row_data)
    print(individual_row_data)
```

we need to put the table data into a dataframe then  combine the columns together with table data

```python
 #displaying the looped data into a dataframe
#length =len(df)
row_df =pd.DataFrame(frame)
row_df
```

```python
#combining both columns and row data
row_df.columns=df.columns
row_df
```

```python
#save the scrapped file in csv format
row_df.to_excel(r"webscrap_companies.xlsx")
```
